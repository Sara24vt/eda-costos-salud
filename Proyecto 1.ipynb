```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd



data = {
    "age": [19, 18, 28, 33, 32, 31, 46, 37, 60, 25],
    "sex": ["female", "male", "male", "male", "male", "female", "female", "female", "female", "male"],
    "bmi": [27.9, 33.77, 33.0, 22.7, 28.88, 25.74, 33.44, 27.74, 25.84, 26.22],
    "children": [0, 1, 3, 0, 0, 0, 1, 3, 0, 0],
    "smoker": ["yes", "no", "no", "no", "no", "no", "no", "no", "no", "no"],
    "region": ["southwest", "southeast", "southeast", "northwest", "northwest", "southwest", "southeast", "northwest", "southwest", "northeast"],
    "charges": [16884.9, 1725.55, 4449.46, 21984.47, 3866.86, 3756.62, 8240.59, 7281.51, 28923.14, 2721.32]
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv("insurance.csv", index=False)
```

```python
plt.figure()
sns.histplot(df["charges"], kde=True)
plt.title("Distribución de costos médicos")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="age", y="charges", data=df)
plt.title("Costos médicos según la edad")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="smoker", y="charges", data=df)
plt.title("Costos médicos según condición de fumador")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="region", y="charges", data=df)
plt.title("Costos médicos por región")
plt.show()
```
